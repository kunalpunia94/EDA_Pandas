```python
#check coorelation
numeric_df = df.select_dtypes(include=['number'])
correlation_matrix = numeric_df.corr()
print(correlation_matrix)
```

```python
sns.heatmap(correlation_matrix,annot=True, cmap="coolwarm", linewidths=1.5, fmt=".2f")
plt.rcParams['figure.figsize']=(12,8)
plt.show()

#another way
#plt.figure(figsize=(20, 6))
#sns.heatmap(correlation_matrix, annot=True, cmap="coolwarm", linewidths=1.5, fmt=".2f")
#plt.show()
```

```python
#group together to look data closer
#used mumeric_only=True gives mean only of numeric values
df.groupby('Continent').mean(numeric_only=True).sort_values(by='2022 Population',ascending=False)
```
